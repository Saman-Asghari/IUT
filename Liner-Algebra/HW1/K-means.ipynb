```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
```

```python
data=np.loadtxt('TrainData.txt', delimiter=',')
```

```python
k = 10
random = np.random.choice(data.shape[0], size=k, replace=False)
initial_centroids = data[random]
kmeans_10_random_choice = KMeans(n_clusters=k, init=initial_centroids, n_init=1, random_state=42)
kmeans_10_random_choice.fit(data)
temp=kmeans_10_random_choice.cluster_centers_
final_centroids=np.where(temp >= 0.4, 1, 0)
fig,axes=plt.subplots(2, 10, figsize=(20, 4))
for i in range(k):
    centroid_image = initial_centroids[i].reshape(40, 30)
    ax = axes[0, i]
    ax.imshow(centroid_image, cmap='gray')
    ax.set_title(f"Init_cent number: {i}")
    ax.axis('off')
for i in range(k):
    centroid_image = final_centroids[i].reshape(40, 30)
    ax = axes[1, i]
    ax.imshow(centroid_image, cmap='gray')
    ax.set_title(f"Fin_cent number: {i}")
    ax.axis('off')

plt.tight_layout()
plt.show()

jaccard_coefficients = []

for i in range(k):
    # Convert centroids to binary format (already done when you created final_centroids)
    initial_centroid_binary = (initial_centroids[i] >= 0.4).astype(int).flatten()
    final_centroid_binary = final_centroids[i].flatten()
    
    # Calculate the Jaccard coefficient
    jc = jaccard_score(initial_centroid_binary, final_centroid_binary, average='binary')
    jaccard_coefficients.append(jc)
    
    print(f"Jaccard coefficient for cluster {i}: {jc}")
```
